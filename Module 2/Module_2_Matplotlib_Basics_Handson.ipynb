# Matplotlib Basics Handson

This notebook demonstrates the most commonly used plotting features and customizations in Matplotlib.

````python
import matplotlib
print("Matplotlib version:", matplotlib.__version__)
````

````python
# -----------------------------------------------------------------------
# 1. IMPORTS & BASIC SETUP
# -----------------------------------------------------------------------
import matplotlib.pyplot as plt
import numpy as np

# %matplotlib inline  # Uncomment if needed in a Jupyter environment

# Sample data for demonstration
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
````

## 2. Basic Plot Types

In this section, we'll explore common plot types:
1. Line Plot
2. Scatter Plot
3. Bar Plot
4. Histogram
5. Box Plot
6. Pie Chart

````python
# -----------------------------------------------------------------------
# 2.1 LINE PLOT
# -----------------------------------------------------------------------
plt.figure(figsize=(6, 4))    # Create a new figure with custom size
plt.plot(x, y, label='sin(x)', color='limegreen', linestyle='-', marker="s", ms=3, linewidth=1)
plt.title("Line Plot of Sine Wave", fontsize=15)
plt.xlabel("X values", fontsize=12)
plt.ylabel("sin(x)", fontsize=12)
plt.legend(loc=3)                  # Show a legend
plt.grid(True)               # Add grid lines for better readability
plt.show()                   # Display the plot
````

````python
# -----------------------------------------------------------------------
# 2.2 SCATTER PLOT
# -----------------------------------------------------------------------
# Generating some random data for demonstration
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c='red', marker='o', alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
````

````python
# -----------------------------------------------------------------------
# 2.3 BAR PLOT
# -----------------------------------------------------------------------
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
````

````python
# -----------------------------------------------------------------------
# 2.4 HISTOGRAM
# -----------------------------------------------------------------------
data = np.random.randn(1000)  # Normal distribution
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()
````

````python
# -----------------------------------------------------------------------
# 2.5 BOX PLOT
# -----------------------------------------------------------------------
# Creating multiple datasets to show box plots side by side
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(5, 2, 100)
data3 = np.random.normal(-3, 1.5, 100)

plt.figure(figsize=(6, 4))
plt.boxplot([data1, data2, data3], tick_labels=['Set 1','Set 2','Set 3'])
plt.title("Box Plot Example")
plt.ylabel("Value")
plt.show()
````

````python
# -----------------------------------------------------------------------
# 2.6 PIE CHART
# -----------------------------------------------------------------------
slices = [40, 25, 20, 15]
labels = ['Slice A', 'Slice B', 'Slice C', 'Slice D']
explode = (0, 0.1, 0, 0)  # Offset the 2nd slice a bit

plt.figure(figsize=(6, 4))
plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title("Pie Chart Example")
plt.show()
````

## 3. Plot Customization

- **Titles, Labels, Legends**
- **Colors, Markers, Line Styles**
- **Figure Size**
- **Axis Limits & Ticks**
- **Grid Lines**

````python
# -----------------------------------------------------------------------
# 3.1 TITLES, LABELS, LEGENDS, COLORS, AND LINE STYLES
# -----------------------------------------------------------------------
# Example with multiple lines, customizing styles
plt.figure(figsize=(6, 4))
plt.plot(x, np.sin(x), label='sin(x)', color='blue', linestyle='-')
plt.plot(x, np.cos(x), label='cos(x)', color='green', linestyle='--')

# Adding title, x/y labels, legend
plt.title("Multiple Lines with Custom Styles")
plt.xlabel("X Axis")
plt.ylabel("Function Value")
plt.legend()
plt.grid(True)
plt.show()
````

````python
# -----------------------------------------------------------------------
# 3.2 AXIS LIMITS AND TICK CUSTOMIZATION
# -----------------------------------------------------------------------
plt.figure(figsize=(6, 4))
plt.plot(x, y, color='red')

# Manually set x and y axis limits
plt.xlim(0, 2*np.pi)
plt.ylim(-1.2, 1.2)

plt.title("Axis Limits and Custom Tick Labels")
# plt.xticks([0, np.pi, 2*np.pi], ['0', '', '2'])  # Custom X tick labels

plt.xticks([0, np.pi, 2*np.pi], ['0', r'$\pi$', r'$2\pi$'])  # Custom X tick labels

plt.yticks([-1, 0, 1])
plt.grid(True)
plt.show()
````

## 4. Working with Multiple Plots

- **Subplots**: Multiple plots within the same figure.
- **Multiple Figures**: Creating separate figures entirely.

````python
# -----------------------------------------------------------------------
# 4.1 SUBPLOTS (SIDE BY SIDE)
# -----------------------------------------------------------------------
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# First subplot
axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title("Sine")

# Second subplot
axes[1].plot(x, np.cos(x), color='green')
axes[1].set_title("Cosine")

plt.suptitle("Subplots Example", fontsize=14)
plt.tight_layout()
plt.show()
````

````python
# -----------------------------------------------------------------------
# 4.2 MULTIPLE FIGURES
# -----------------------------------------------------------------------
# Figure 1
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)')
plt.title("Figure 1")
plt.legend()
plt.show()

# Figure 2
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c='orange')
plt.title("Figure 2")
plt.show()
````

## 5. Styling and Themes

Matplotlib has built-in style sheets (e.g., `ggplot`, `seaborn`, `bmh`) and can also use custom styles. You can set a style using:

```python
plt.style.use('ggplot')

Also, you can annotate your plots to call attention to specific points.

````python
# -----------------------------------------------------------------------
# 5.1 USING A BUILT-IN STYLE AND ANNOTATIONS
# -----------------------------------------------------------------------
plt.style.use('ggplot')  # Switch to ggplot style

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)')

# Annotate a specific point
x_point = np.pi
y_point = np.sin(x_point)
plt.scatter(x_point, y_point, color='black')
plt.annotate('Value at ', xy=(x_point, y_point),
             xytext=(x_point+0.5, y_point),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title("Styled Plot with Annotations")
plt.xlabel("X")
plt.ylabel("Y = sin(x)")
plt.legend()
plt.show()
````

````python
# Revert to default style for subsequent plots
plt.style.use('default')
````

## 6. Saving and Exporting Plots

You can save your figure using:
```python
plt.savefig('my_figure.png', dpi=300, bbox_inches='tight')
where:

dpi sets the resolution.
bbox_inches='tight' ensures minimal whitespace.

````python
# -----------------------------------------------------------------------
# 6.1 SAVING A PLOT
# -----------------------------------------------------------------------
plt.figure(figsize=(6, 4))
plt.plot(x, np.sin(x), color='blue')
plt.title("Plot to be Saved")

# Uncomment this line to actually save the figure locally
plt.savefig('my_plot.png', dpi=300, bbox_inches='tight')

plt.show()
````

## 7. Tips, Tricks, and Good Practices

1. **Use Subplots** for multiple visualizations rather than stacking multiple plots.
2. **Label axes and add a legend** for clarity.
3. **Use styles** and color maps to create visually appealing plots.
4. **Keep track of figure and axes** objects with the OOP interface (`fig, ax = plt.subplots()`) for more complex layouts.
5. **Check official Matplotlib documentation** for advanced features (3D plots, interactive backends, etc.).

## 8. Practical Example: Combining Plot Types

In this brief example, we'll show how to overlay a scatter plot on a line plot to compare two related data views.

````python
# -----------------------------------------------------------------------
# 8.1 PRACTICAL EXAMPLE
# -----------------------------------------------------------------------
# Synthetic data
time = np.linspace(0, 10, 50)
values_line = 2 * time + 1
values_scatter = values_line + np.random.normal(0, 3, size=time.size)

plt.figure(figsize=(7, 5))
plt.plot(time, values_line, label='Ideal Trend', color='darkblue', linewidth=2)
plt.scatter(time, values_scatter, label='Observed Data', color='red', alpha=0.7)

plt.title("Line + Scatter Plot Example")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()


              
````
